# 10. Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score,recall_score,f1_score
from sklearn.metrics import classification_report
from sklearn.metrics import precision_recall_curve, auc, average_precision_score
import matplotlib.pyplot as plt
import seaborn as sns
log = LogisticRegression()
log.fit(X_train,y_train)
```

```python
y_pred1 = log.predict(X_test)
```

```python
accuracy_score(y_test,y_pred1)
```

```python
precision_score(y_test,y_pred1)
```

```python
recall_score(y_test,y_pred1)
```

```python
f1_score(y_test,y_pred1)
```

```python
print("Classification Report (Train Set):\n", classification_report(y_train, y_train))
print("Classification Report (Test Set):\n", classification_report(y_test, y_pred1))
```

```python
precision, recall, _ = precision_recall_curve(y_test, y_pred1)
auprc = auc(recall, precision)
average_precision = average_precision_score(y_test, y_pred1)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(recall, precision, label=f'AUPRC = {auprc:.2f}\nAvg Precision = {average_precision:.2f}')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend(loc='best')
plt.show()
```
